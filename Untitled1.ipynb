```python
import cv2
```

```python
model=cv2.CascadeClassifier("/content/haarcascade_frontalface_default.xml")
```

```python
faces=model.detectMultiScale(img)
```

```python
faces
```

```python
faces.shape
```

```python
for face in faces:
  x,y,w,h=face

  img=cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
```

```python
plt.imshow(img)
```
